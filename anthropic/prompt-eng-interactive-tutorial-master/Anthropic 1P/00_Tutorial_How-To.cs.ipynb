### Anthropic SDK a Messages API
V celém tutorialu budeme používat [Anthropic python SDK](https://docs.anthropic.com/claude/reference/client-sdks) a [Messages API](https://docs.anthropic.com/claude/reference/messages_post).

Níže je příklad, jak bude v tomto tutorialu vypadat spuštění promptu. Nejprve vytvoříme `get_completion`, pomocnou funkci, která pošle prompt Claude a vrátí odpověď vygenerovanou Claude. Tuto buňku teď spusťte.


```python
import anthropic

client = anthropic.Anthropic(api_key=API_KEY)

def get_completion(prompt: str):
    message = client.messages.create(
        model=MODEL_NAME,
        max_tokens=2000,
        temperature=0.0,
        messages=[
          {"role": "user", "content": prompt}
        ]
    )
    return message.content[0].text
```

Teď napíšeme ukázkový prompt pro Claude a vypíšeme výstup Claude spuštěním pomocné funkce `get_completion`. Spuštění buňky níže vypíše odpověď Claude pod buňkou.

Klidně si pohrajte s řetězcem promptu a sledujte, jak se mění odpovědi Claude.


```python
# Prompt
prompt = "Hello, Claude!"

# Get Claude's response
print(get_completion(prompt))
```
