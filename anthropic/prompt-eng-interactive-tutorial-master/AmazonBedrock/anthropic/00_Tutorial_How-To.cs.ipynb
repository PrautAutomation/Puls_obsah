### Anthropic SDK a Messages API
V celem tutorialu budeme pouzivat [Anthropic python SDK](https://docs.anthropic.com/claude/reference/claude-on-amazon-bedrock) a [Messages API](https://docs.anthropic.com/claude/reference/messages_post).

Nize je priklad toho, jak bude v tomto tutorialu vypadat spusteni promptu.

Nejprve nastavime a ulozime nazev modelu a region.

```python
import boto3
session = boto3.Session() # create a boto3 session to dynamically get and set the region name
AWS_REGION = session.region_name
print("AWS Region:", AWS_REGION)
MODEL_NAME = "anthropic.claude-3-haiku-20240307-v1:0"

%store MODEL_NAME
%store AWS_REGION
```

Potom vytvorime `get_completion`, pomocnou funkci, ktera posle prompt Claudovi a vrati jeho vygenerovanou odpoved. Tuto bunku ted spustte.

```python
from anthropic import AnthropicBedrock

client = AnthropicBedrock(aws_region=AWS_REGION)

def get_completion(prompt, system=''):
    message = client.messages.create(
        model=MODEL_NAME,
        max_tokens=2000,
        temperature=0.0,
        messages=[
          {"role": "user", "content": prompt}
        ],
        system=system
    )
    return message.content[0].text
```

Teď napíšeme ukázkový prompt pro Claude a vypíšeme výstup Claude spuštěním pomocné funkce `get_completion`. Spuštění buňky níže vypíše odpověď Claude pod buňkou.

Klidně si pohrajte s řetězcem promptu a sledujte, jak se mění odpovědi Claude.


```python
# Prompt
prompt = "Hello, Claude!"

# Get Claude's response
print(get_completion(prompt))
```

Promenne `MODEL_NAME` a `AWS_REGION` definovane vyse se budou pouzivat v celem tutorialu. Jen dbejte na to, abyste bunky na kazde strance tutorialu spousteli shora dolu.
