# Dodatek 10.3: Search & Retrieval

Věděli jste, že můžete Claude použít k tomu, aby za vás **vyhledával ve Wikipedii**? Claude dokáže najít a načíst články, které potom může shrnovat, syntetizovat, použít k psaní nového obsahu a mnohem víc. A nejde jen o Wikipedii. Vyhledávat můžete i ve vlastních dokumentech, ať už jsou uložené jako prostý text, nebo vložené do vektorového úložiště.

Podívejte se na naše [RAG cookbook examples](https://github.com/anthropics/anthropic-cookbook/blob/main/third_party/Wikipedia/wikipedia-search-cookbook.ipynb), kde se naučíte, jak doplnit znalosti Claude a zlepšit přesnost a relevanci odpovědí pomocí dat získaných z vektorových databází, Wikipedie, internetu a dalších zdrojů. Najdete tam také informace o použití některých [embeddingů](https://docs.anthropic.com/claude/docs/embeddings) a nástrojů pro vektorové databáze.

Pokud vás zajímají pokročilé RAG architektury s Claude, podívejte se na [Claude 3 technical presentation slides on RAG architectures](https://docs.google.com/presentation/d/1zxkSI7lLUBrZycA-_znwqu8DDyVhHLkQGScvzaZrUns/edit#slide=id.g2c736259dac_63_782).
