# Priloha 10.4: Vyhledavani a nacitani informaci

Vedeli jste, ze Claude muzete pouzit k tomu, aby za vas **prohledaval Wikipedii**? Claude umi najit a nacist clanky; potom ho muzete pouzit i k jejich shrnuti a syntetizovani, k psani noveho obsahu z nalezenych informaci a k mnoha dalsim vecem. A nejde jen o Wikipedii. Muzete prohledavat i vlastni dokumenty, at uz jsou ulozene jako prosty text, nebo vlozene ve vektorovem ulozisti.

Podivejte se na nase [RAG cookbook priklady](https://github.com/anthropics/anthropic-cookbook/blob/main/third_party/Wikipedia/wikipedia-search-cookbook.ipynb), kde se dozvite, jak doplnit Claudovy znalosti a zlepsit presnost a relevanci jeho odpovedi pomoci dat nactenych z vektorovych databazi, Wikipedie, internetu a dalsich zdroju. Najdete tam take informace o pouziti vybranych [embeddingu](https://docs.anthropic.com/claude/docs/embeddings) a nastroju pro vektorove databaze.

Pokud vas zajimaji pokrocile RAG architektury s Claudem, projdete si nase [technicke prezentacni slidy ke Claude 3 o RAG architekturach](https://docs.google.com/presentation/d/1zxkSI7lLUBrZycA-_znwqu8DDyVhHLkQGScvzaZrUns/edit#slide=id.g2c736259dac_63_782).
