## Nastavení

Spusťte následující nastavovací buňku, která načte API klíč a vytvoří pomocnou funkci `get_completion`.


```python
%pip install anthropic

# Import python's built-in regular expression library
import re
import anthropic

# Retrieve the API_KEY & MODEL_NAME variables from the IPython store
%store -r API_KEY
%store -r MODEL_NAME

client = anthropic.Anthropic(api_key=API_KEY)

def get_completion(prompt: str, system_prompt=""):
    message = client.messages.create(
        model=MODEL_NAME,
        max_tokens=2000,
        temperature=0.0,
        system=system_prompt,
        messages=[
          {"role": "user", "content": prompt}
        ]
    )
    return message.content[0].text
```

## Lekce

Navážeme na téma, že Claude nemá žádný kontext kromě toho, co mu řeknete. Někdy je důležité **vyzvat Claude, aby převzal konkrétní roli včetně potřebného kontextu**. Tomu se říká role prompting. Čím detailnější kontext role dáte, tím lépe.

**Nastavení role může zlepšit výkon Claude** v různých oblastech, od psaní přes programování až po shrnování. Je to podobné, jako když člověku pomůže instrukce `přemýšlej jako ______`. Role prompting může také změnit styl, tón a způsob odpovědi Claude.

**Poznámka:** Role prompting může být buď v system promptu, nebo jako součást uživatelské zprávy.


### Příklady

V příkladu níže vidíme, že bez role promptingu Claude poskytne **přímočarou a nestylizovanou odpověď**, když ho požádáme o jednu větu s pohledem na skateboarding.

Když ale Claude nastavíme do role kočky, jeho perspektiva se změní, a tím se **tón, styl i obsah odpovědi přizpůsobí nové roli**.

**Poznámka:** Bonusová technika je **dodat Claude kontext o zamýšleném publiku**. Níže bychom mohli prompt upravit tak, aby Claude věděl, ke komu mluví. `You are a cat` povede k jiné odpovědi než `you are a cat talking to a crowd of skateboarders`.

Zde je prompt bez role promptingu v system promptu:


```python
# Prompt
PROMPT = "In one sentence, what do you think about skateboarding?"

# Print Claude's response
print(get_completion(PROMPT))
```

Zde je stejná uživatelská otázka, ale tentokrát s role promptingem.


```python
# System prompt
SYSTEM_PROMPT = "You are a cat."

# Prompt
PROMPT = "In one sentence, what do you think about skateboarding?"

# Print Claude's response
print(get_completion(PROMPT, SYSTEM_PROMPT))
```

Role prompting můžete použít k tomu, aby Claude napodobil určitý styl psaní, mluvil určitým hlasem nebo upravil složitost odpovědí. **Role prompting může také zlepšit výkon Claude v matematických nebo logických úlohách.**

Například v příkladu níže existuje jednoznačně správná odpověď, a tou je ano. Claude se ale splete a myslí si, že mu chybí informace, i když nechybí:


```python
# Prompt
PROMPT = "Jack is looking at Anne. Anne is looking at George. Jack is married, George is not, and we don’t know if Anne is married. Is a married person looking at an unmarried person?"

# Print Claude's response
print(get_completion(PROMPT))
```

Co když ale **nastavíme Claude, aby jednal jako logický bot**? Jak se tím změní jeho odpověď?

Ukáže se, že s touto novou rolí Claude odpoví správně. I když ne nutně ze všech správných důvodů.


```python
# System prompt
SYSTEM_PROMPT = "You are a logic bot designed to answer complex logic problems."

# Prompt
PROMPT = "Jack is looking at Anne. Anne is looking at George. Jack is married, George is not, and we don’t know if Anne is married. Is a married person looking at an unmarried person?"

# Print Claude's response
print(get_completion(PROMPT, SYSTEM_PROMPT))
```

**Poznámka:** V průběhu kurzu uvidíte, že existuje **mnoho technik prompt engineeringu, které mohou vést k podobným výsledkům**. Které techniky použijete, je na vás a vašich preferencích. Doporučujeme **experimentovat a najít si vlastní styl prompt engineeringu**.

### Cvičení 3.1 - Oprava matematiky
V některých případech může mít **Claude potíže s matematikou**, i jednoduchou. Níže Claude nesprávně vyhodnotí matematický problém jako správně vyřešený, i když je ve druhém kroku zjevná aritmetická chyba. Všimněte si, že Claude chybu zachytí při postupu krok za krokem, ale nedojde k závěru, že celé řešení je špatně.

Upravte `PROMPT` a/nebo `SYSTEM_PROMPT` tak, aby Claude ohodnotil řešení jako vyřešené `incorrectly`, ne jako správně vyřešené.



```python
# System prompt - if you don't want to use a system prompt, you can leave this variable set to an empty string
SYSTEM_PROMPT = ""

# Prompt
PROMPT = """Is this equation solved correctly below?

2x - 3 = 9
2x = 6
x = 3"""

# Get Claude's response
response = get_completion(PROMPT, SYSTEM_PROMPT)

# Function to grade exercise correctness
def grade_exercise(text):
    if "incorrect" in text or "not correct" in text.lower():
        return True
    else:
        return False

# Print Claude's response and the corresponding grade
print(response)
print("\n--------------------------- GRADING ---------------------------")
print("This exercise has been correctly solved:", grade_exercise(response))
```
